```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32,(3,3), activation='relu',
                       input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(10,activation='sigmoid'))
```

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['acc'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir, #타깃 디렉터리
                                                    target_size=(150,150), #모든 이미지 150X150으로 변경
                                                    batch_size=20, #
                                                    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(validation_dir, target_size=(150,150),batch_size=20, class_mode='categorical')
```

```python
from keras import models

layer_outputs = [layer.output for layer in model_2.layers[:8]] #상위 8개 층에 대한 출력을 리스트로 저장함
activation_model = models.Model(inputs=model_2.input, outputs=layer_outputs) #입력에 대해 8개의 층의 출력을 반환하는 모델
```

```python
activations =activation_model.predict(img_tensor)
#층의 활성화 마다 하나씩 즉, 각 층에 해당하는 특성맵을 가진 8개 넘파이배열로 이루어진 리스트를 반환
```

```python
first_layer_activation= activations[0]
print(first_layer_activation.shape) # (1,148,148,32)  150X150의 입력값이 3X3의 윈도우를 거쳐 나온 148X148특성맵
```

```python
plt.matshow(first_layer_activation[0, :, :, 16], cmap='viridis')
#1번째 층의 16번째 특성맵
```

```python
layer_names = [] #각 층의 이름을 그래프의 이름으로 저장하기 위한 리스트

for layer in model_2.layers[:8]: #각층마다 이름을 가져와 append함
    layer_names.append(layer.name)

    
images_per_row = 16 #가로 16개 이미지를 나열하기 위함

for layer_name, layer_activation in zip(layer_names, activations): #각 층의 이름과 그 층에 해당하는 특성맵들을 묶음
    n_features = layer_activation.shape[-1] #(1,148,148,32) 는 (이미지개수, 위아래 픽셀, 특성맵개수) 중 -1에 해당하는 특성맵 개수를 추출 
    
    size = layer_activation.shape[1] #size는 픽셀의 148 해당
    
    n_cols = n_features // images_per_row # 32개 특성맵 // 16개 가로나열
    display_grid = np.zeros((size*n_cols, images_per_row*size)) # 148*(32//16) 행 , 16*148 열  (2행 16열)
    
    for col in range(n_cols): #2행
        for row in range(images_per_row): #16열
            channel_image = layer_activation[0, :, :, col*images_per_row + row] #첫 이미지에 대해 0,0 을 기준으로 → X16 ,  ↓X1, → X16 
            channel_image -= channel_image.mean() #0~255사이의 픽셀로 표현하기위함. 평균을 빼고 
            channel_image /= channel_image.std() # 표준편차로 나누고 ( -2.0~2.0 값들이 생성됨)
            channel_image *= 64 # -128 ~+128 사이 값에 놓이도록 증폭한 후
            channel_image +=128  #0~255로 만들기 위해 128을 더하고
            channel_image = np.clip(channel_image, 0, 255).astype('uint8') #클리핑 0과 255사이값이 아니면 넘어버린 값을 최소값 혹은 최대값으로 만들어줌
            display_grid[col*size : (col+1) * size,   row*size : (row+1) * size] = channel_image
                       #      0~1*148 : 1~2*148    ,      0~16*148 : 1~17*148
    
    
    scale = 1./size # 1/148
    
    plt.figure(figsize=(scale * display_grid.shape[1], scale* display_grid.shape[0])) #일정 범위 안에 일정 갯수 사진을 넣기 위함
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
    
plt.show()
```

```python
#필터시각화를 위한 손실 텐서 정의

from keras import backend as K

model = load_model('number_recog.h5')

model.summary()
```

```python
layer_name = 'block3_conv1' #네트워크중 block3_conv1 층 필터들에서
filter_index = 0  #0번째 인덱스에 해당하는 필터
```

```python
#텐서를 이미지형태로 변환하기 위한 유틸리티함수
#(1,150,150,3) 크기의 결과값은 0~255사이의 정수가 아니므로 이미지로 나타내기 위한 후처리하는 과정임
def deprocess_image(x):
    x-=x.mean() 
    x/=(x.std()+1e-5)
    x*=0.1 #텐서 평균이 0, 표준편차가 0.1이 되도록  정규화
    
    x+=0.5
    x=np.clip(x,0,1) #[0,1]로 클리핑함
    
    x*=255 #rgb값으로 변환
    x=np.clip(x,0,255).astype('uint8')
    return x
```

```python
#위 코드를 모아 함수로 만듬
def generate_pattern(layer_name, filter_index, size=150):
    layer_output = model.get_layer(layer_name).output#그 모델의 해당 층에 해당하는 정보들을 가져옴
    loss = K.mean(layer_output[:, :, :, filter_index]) # 0번째 필터의 평균값
    
    grads = K.gradients(loss, model.input)[0] #손실에 대한 입력이미지의 그래디언트 계산
    grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5) #그래디언트 정규화
    
    iterate = K.function([model.input],[loss, grads]) ##입력 이미지에 대한 손실과 그래디언트를 반환
    input_img_data = np.random.random((1,size,size,3))*20 + 128. #잡음이 섞인 회색 이미지로 시작
    
    step = 1. #업데이트할 그래디언트의 크기
    for i in range(40): #경사상승법을 40회 실행
        loss_value, grads_value = iterate([input_img_data]) #손실과 그래디언트 계산 

        input_img_data += grads_value * step #손실을 최대화 하는 방향으로 입력이미지 수정  

    img = input_img_data[0]
    return deprocess_image(img)
```

```python
plt.imshow(generate_pattern('conv2d_12',0))
```

```python
layer_name = 'conv2d_12'
size = 64
margin = 5

results=np.zeros((8*size + 7*margin, 8*size + 7*margin, 3),dtype='uint8')#결과를 담을 빈 이미지 

for i in range(8): #행
    for j in range(8): #열
        filter_img = generate_pattern(layer_name, i+(j*8), size=size) #layer_name에 있는 i+(j*8)번째 필터에 대한 패턴을 생성
        
        horizontal_start = i*size + i*margin 
        horizontal_end = horizontal_start + size
        vertical_start = j*size + j*margin
        vertical_end = vertical_start + size
        
        results[horizontal_start : horizontal_end, vertical_start : vertical_end, :] = filter_img
        
plt.figure(figsize=(20, 20))
plt.imshow(results)
```
